```python
def init_params():
    W1 = np.random.rand(32, 784) - 0.5
    b1 = np.random.rand(32, 1) - 0.5

    W2 = np.random.rand(32, 32) - 0.5
    b2 = np.random.rand(32, 1) - 0.5

    W3 = np.random.rand(32, 32) - 0.5
    b3 = np.random.rand(32, 1) - 0.5

    W4 = np.random.rand(32, 32) - 0.5
    b4 = np.random.rand(32, 1) - 0.5

    W5 = np.random.rand(10, 32) - 0.5
    b5 = np.random.rand(10, 1) - 0.5

    return W1, b1, W2, b2, W3, b3, W4, b4, W5, b5

def Relu(Z):
    return np.maximum(0,Z)  

def Softmax(Z):
    return np.exp(Z)/sum(np.exp(Z))

def forward_prop(W1, b1, W2, b2, W3, b3, W4, b4, W5, b5, X):
    Z1 = W1.dot(X) + b1
    A1 = Relu(Z1)

    Z2 = W2.dot(A1) + b2
    A2 = Relu(Z2)

    Z3 = W3.dot(A2) + b3
    A3 = Relu(Z3)

    Z4 = W4.dot(A3) + b4
    A4 = Relu(Z4)

    Z5 = W5.dot(A4) + b5
    A5 = Softmax(Z5)

    return Z1, A1, Z2, A2, Z3, A3, Z4, A4, Z5, A5

def one_hot(Y):
    one_hot_Y = np.zeros((Y.size, Y.max()+1))
    one_hot_Y[np.arange(Y.size),Y] =1
    one_hot_Y = one_hot_Y.T
    return one_hot_Y

def deriv_relu(Z):
    return Z>0

def back_prop(Z1, A1, Z2, A2, Z3, A3, Z4, A4, Z5, A5, W2, W3, W4, W5, X, Y):
    one_hot_Y = one_hot(Y)

    dL_Z5 = A5 - one_hot_Y
    dL_W5 = 1/m * dL_Z5.dot(A4.T)
    dL_b5 = 1/m * np.sum(dL_Z5)

    dL_Z4 = W5.T.dot(dL_Z5) * deriv_relu(Z4)
    dL_W4 = 1/m * dL_Z4.dot(A3.T)
    dL_b4 = 1/m * np.sum(dL_Z4)

    dL_Z3 = W4.T.dot(dL_Z4) * deriv_relu(Z3)
    dL_W3 = 1/m * dL_Z3.dot(A2.T)
    dL_b3 = 1/m * np.sum(dL_Z3)

    dL_Z2 = W3.T.dot(dL_Z3) * deriv_relu(Z2)
    dL_W2 = 1/m * dL_Z2.dot(A1.T)
    dL_b2 = 1/m * np.sum(dL_Z2)

    dL_Z1 = W2.T.dot(dL_Z2) * deriv_relu(Z1)
    dL_W1 = 1/m * dL_Z1.dot(X.T)
    dL_b1 = 1/m * np.sum(dL_Z1)

    return dL_W1, dL_b1, dL_W2, dL_b2, dL_W3, dL_b3, dL_W4, dL_b4, dL_W5, dL_b5

def update_params(W1, b1, W2, b2, W3, b3, W4, b4, W5, b5, dL_W1, dL_b1, dL_W2, dL_b2, dL_W3, dL_b3, dL_W4, dL_b4, dL_W5, dL_b5, alpha):
    W1 = W1 - alpha * dL_W1
    b1 = b1 -  alpha * dL_b1
    W2 = W2 - alpha * dL_W2
    b2 = b2 -  alpha * dL_b2
    W3 = W3 - alpha * dL_W3
    b3 = b3 -  alpha * dL_b3
    W4 = W4 - alpha * dL_W4
    b4 = b4 -  alpha * dL_b4
    W5 = W5 - alpha * dL_W5
    b5 = b5 -  alpha * dL_b5
    return W1, b1, W2, b2, W3, b3, W4, b4, W5, b5
```

```python
def get_predicts(A2):
    return np.argmax(A2,0)

def get_accuracy(predictions, Y):
    print(predictions, Y)
    return np.sum(predictions==Y) / Y.size

def gradient_descent(X,Y, iters, alpha):
    W1, b1, W2, b2, W3, b3, W4, b4, W5, b5 = init_params()
    for i in range(iters):
        Z1, A1, Z2, A2, Z3, A3, Z4, A4, Z5, A5 = forward_prop(W1, b1, W2, b2, W3, b3, W4, b4, W5, b5, X)
        dL_W1, dL_b1, dL_W2, dL_b2, dL_W3, dL_b3, dL_W4, dL_b4, dL_W5, dL_b5 = back_prop(Z1, A1, Z2, A2, Z3, A3, Z4, A4, Z5, A5, W2, W3, W4, W5, X, Y)
        W1, b1, W2, b2, W3, b3, W4, b4, W5, b5 = update_params(W1, b1, W2, b2, W3, b3, W4, b4, W5, b5, dL_W1, dL_b1, dL_W2, dL_b2, dL_W3, dL_b3, dL_W4, dL_b4, dL_W5, dL_b5, alpha)
        if(i%10==0):
            print("Interation: ", i)
            print("Accuracy: ", get_accuracy(get_predicts(A2), Y))
    return W1, b1, W2, b2, W3, b3, W4, b4, W5, b5
```
